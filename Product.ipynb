```python
tech_products = []


def add_product():
    product = input("Enter the tech product you want to add: ")
    tech_products.append(product)
    print(f"'{product}' has been added successfully.")


def remove_product():
    product = input("Enter the tech product you want to remove: ")
    if product not in tech_products:
        print("This product is not available in the store.")
    else:
        tech_products.remove(product)
        print(f"'{product}' has been removed from the store.")


def show_products():
    if len(tech_products) == 0:
        print("The store is currently empty.")
    else:
        print("Tech products available in the store:")
        for p in tech_products:
            print(f"- {p}")
        print(f"Total products: {len(tech_products)}")


def edit_product():
    old = input("Enter the tech product you want to edit: ")
    if old not in tech_products:
        print("This product is not available in the store.")
    else:
        new = input("Enter the new name for the product: ")
        index = tech_products.index(old)
        tech_products[index] = new
        print(f"'{old}' has been updated to '{new}'.")


def exit_program():
    print("Thank you for shopping with us. Goodbye.")
    return False


while True:
    print("""
Choose an option:
1 - Add Product
2 - Remove Product
3 - Show Products
4 - Edit Product
5 - Exit
""")
    choice = input("Enter your choice: ")

    if choice == "1":
        add_product()
    elif choice == "2":
        remove_product()
    elif choice == "3":
        show_products()
    elif choice == "4":
        edit_product()
    elif choice == "5":
        break
    else:
        print("Invalid choice. Please try again.")
```
